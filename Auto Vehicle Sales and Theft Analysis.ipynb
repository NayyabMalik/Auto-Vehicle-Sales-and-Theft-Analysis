```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv(r"C:\Users\PMLS\Downloads\dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Series Display Name'].unique()
```

```python
df['date']=pd.to_datetime(df['Observation Date'])
```

```python

# Assuming df is your DataFrame with a 'date' column
# Create a new DataFrame with year, day, week, and quarter extracted from the 'date' column
new_df = pd.DataFrame({
    'year': df['date'].dt.year,
    'day': df['date'].dt.day,
    'week': df['date'].dt.isocalendar().week,
    'quarter': df['date'].dt.quarter,
    'hour':df['date'].dt.hour
})
```

```python
combined_df = pd.concat([df, new_df], axis=1)
```

```python
combined_df.head()
```

```python
col=['Unit','Observation Status Comment','Observation Date','Dataset Name','date','Series Key','Observation Status']
combined_df.drop(columns=col,inplace=True)
```

```python
combined_df.info()
```

```python
df['Series Display Name'].unique()
```

```python
df['Sequence No.'].unique()
```

```python
df['Series name'].value_counts()
```

```python
combined_df.drop(columns=['Series Display Name','Sequence No.'],inplace=True)
```

```python


motor_cycles_production_df = combined_df[combined_df['Series name'] == 'Production of Motor cycles']

production_by_year_quarter = motor_cycles_production_df.groupby(['year', 'quarter'])['Observation Value'].sum().reset_index()

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(data=production_by_year_quarter, x='quarter', y='Observation Value', hue='year')
plt.title('Production of Motor Cycles by Year and Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Production')
plt.legend(title='Year')
plt.show()
```

```python

# Filter the DataFrame for 'Production of Motor cycles'
motor_cycles_sales_df = combined_df[combined_df['Series name'] == 'Sales of Motor cycles']

# Group by year and sum the production
production_by_year = motor_cycles_sales_df.groupby('year')['Observation Value'].sum().reset_index()

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(data=production_by_year, x='year', y='Observation Value')
plt.title('Sales of Motor Cycles by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```
